```python
### Your code here
## Feel free to use multiple cells
df = pd.read_excel("default of credit card clients.xls", 
                   sheet = 0, skiprows= 1, header = 0)

#Drop the ID column as it is not needed
df = df.drop('ID', axis=1)

#PAY_1 is missing so we considrer PAY_0 as Pay_1
df = df.rename(index=str, columns={"PAY_0": "PAY_1"})

#We can go ahread and remove features BILL_AMT and PAY_AMT as we already have that information in categorical features PAY_*
import re
pattern = re.compile("^BILL_AMT[0-9]+$")
bill_columns = [ x for x in df.columns if (pattern.match(x))]
df = df.drop(bill_columns, axis=1)

pattern1 = re.compile("^PAY_AMT[0-9]+$")
pay_columns = [ x for x in df.columns if (pattern1.match(x))]
df = df.drop(pay_columns, axis=1)


df.head(5)
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.2, random_state=123)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

# A class to select numerical or categorical columns 
# since Scikit-Learn doesn't handle DataFrames yet
class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names]
```

```python
#Pipeline for numerical columns
from sklearn.pipeline import Pipeline

num_pipeline = Pipeline([
        ("select_numeric", DataFrameSelector(["LIMIT_BAL", "AGE"])),
    ])

num_pipeline.fit_transform(train_set)
```

```python
cat_pipeline = Pipeline([
        ("select_cat", DataFrameSelector(["SEX", "EDUCATION", "MARRIAGE", "PAY_1", "PAY_2", "PAY_3", 
                                          "PAY_4", "PAY_5", "PAY_6"])),
    ])
cat_pipeline.fit_transform(train_set)
```

```python
#Merge the two pipelines
from sklearn.pipeline import FeatureUnion
preprocess_pipeline = FeatureUnion(transformer_list=[
        ("num_pipeline", num_pipeline),
        ("cat_pipeline", cat_pipeline),
    ])
```

```python
X_train = preprocess_pipeline.fit_transform(train_set)
```

```python
y_train = train_set["default payment next month"]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_curve
```

```python
param_grid = {
    'n_estimators': [4, 5, 10, 20, 50]
}
rf = RandomForestClassifier(random_state=123)
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, cv = 5, scoring="precision")
grid_search.fit(X_train, y_train)
grid_search.best_params_,grid_search.best_score_
```

```python
#Running with n_neighbors {3, 5, 10, 20} 
param_grid = {
    'n_neighbors': [3, 5, 10, 20]
}
knn = KNeighborsClassifier()
grid_search = GridSearchCV(estimator = knn, param_grid = param_grid, cv = 5, scoring="precision")
grid_search.fit(X_train, y_train)
grid_search.best_params_,grid_search.best_score_
```

```python
#ROC for Training data using RandomForestClassifier and best n_estimator 50
forest_clf = RandomForestClassifier(random_state=123, n_estimators=50)
y_probas_forest = cross_val_predict(forest_clf, X_train, y_train, cv=5, method="predict_proba")
y_scores_forest = y_probas_forest[:, 1] # score = proba of positive class
fpr_forest, tpr_forest, thresholds_forest = roc_curve(y_train,y_scores_forest)
```

```python
#ROC for Training data using KNN Classifier and best n_neighbour 10
knn_clf = KNeighborsClassifier(n_neighbors=20)
y_probas_knn = cross_val_predict(knn_clf, X_train, y_train, cv=5, method="predict_proba")
y_scores_knn = y_probas_knn[:, 1] # score = proba of positive class
fpr_knn, tpr_knn, thresholds_knn = roc_curve(y_train,y_scores_knn)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([0, 1, 0, 1])
    plt.xlabel('False Positive Rate', fontsize=16)
    plt.ylabel('True Positive Rate', fontsize=16)
```

```python
plt.figure(figsize=(8, 6))
plot_roc_curve(fpr_knn, tpr_knn, "KNN")
plot_roc_curve(fpr_forest, tpr_forest, "Random Forest")
plt.legend(loc="lower right", fontsize=16)
plt.show()
```
